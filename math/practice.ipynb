```python
import numpy as np

A = np.array([
    [1, 3, 4],
    [3, 3, 1],
    [4, 1, 2]
])

# Находим собственные значения и собственные векторы
eigvals, eigvecs = np.linalg.eigh(A)

print("Собственные значения:")
print(eigvals)

print("\nМатрица собственных векторов (столбцы — нормированные собственные векторы):")
print(eigvecs)

# Проверка разложения A = V Λ V^T
Lambda = np.diag(eigvals)
A_reconstructed = eigvecs @ Lambda @ eigvecs.T

print("\nПроверка: восстановленная матрица A:")
print(A_reconstructed)
```
